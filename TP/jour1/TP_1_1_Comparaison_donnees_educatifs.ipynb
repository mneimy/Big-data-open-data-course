```python
from pathlib import Path

def resolve_data_dir() -> Path:
    """Racine du dépôt = dossier qui contient `donnees/`. Lancer Jupyter depuis cette racine."""
    cwd = Path.cwd()
    if (cwd / "donnees").is_dir():
        return cwd
    p = cwd
    for _ in range(4):
        if (p / "donnees").is_dir():
            return p
        p = p.parent
    return cwd

ROOT = resolve_data_dir()
DATA = ROOT / "donnees"
print("Répertoire données :", DATA.resolve())
```

```python
# Renommer un fichier
import pandas as pd
from pathlib import Path

csv_path = DATA / "fr-en-annuaire-education.csv"
if not csv_path.exists():
    # Cherche un fichier similaire à fr-en-annuaire-education*.csv
    similar_files = list(DATA.glob("fr-en-annuaire-education*.csv"))
    # Exclure le nom recherché si présent
    similar_files = [f for f in similar_files if f.name != "fr-en-annuaire-education.csv"]
    if similar_files:
        # On renomme le premier fichier trouvé
        similar_file = similar_files[0]
        print(f"Renommage automatique de {similar_file.name} en fr-en-annuaire-education.csv")
        similar_file.rename(csv_path)
    else:
        raise FileNotFoundError(
            f"Fichier absent : {csv_path}. Aucun fichier similaire trouvé. Téléchargez l'export (voir cellule markdown ci-dessus 1 - Téléchargement)."
        )
```

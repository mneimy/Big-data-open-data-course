# TP 2.2 — Démonstration Hadoop / HDFS

Ce notebook **guide** la manipulation HDFS sur la **VM de formation** : Hadoop est **installé localement** (pas de Docker). Les commandes s’exécutent dans **PowerShell** ou l’**invite de commandes**, ou en cellules `!` dans Jupyter si le client `hdfs` est dans le `PATH`.

**Donnée volumineuse conseillée** : effectifs VP/BTS — voir `Docs/sources_data.md` section TP 2.2 (fichier placé dans `donnees\` à la racine du dépôt).

**Références** : `README.md` (section Hadoop / HDFS).

## 1. Vérifier et démarrer le cluster Hadoop

Ouvrir un terminal **à la racine du dépôt** (ou n’importe où si `hdfs` est dans le `PATH`).

**Vérifier que Hadoop répond** :

```powershell
hdfs dfs -ls /
jps
```

`jps` doit lister au minimum **NameNode** et **DataNode** (et éventuellement ResourceManager / NodeManager si YARN est démarré).

**Si le cluster n’est pas démarré** (erreur de connexion au NameNode), lancer les services depuis le dossier `sbin` de Hadoop (adapter `%HADOOP_HOME%` si besoin) :

```powershell
%HADOOP_HOME%\sbin\start-dfs.cmd
%HADOOP_HOME%\sbin\start-yarn.cmd
```

Puis revérifier : `hdfs dfs -ls /` et `jps`.

Interface web du NameNode : **http://localhost:9870**

## 2. Copier un fichier local vers HDFS

Télécharger le CSV des effectifs (voir `Docs/sources_data.md`) dans `donnees\` à la racine du projet, puis l’envoyer sur HDFS.

Remplacer `formation` par votre identifiant Windows si différent, et adapter le nom du fichier CSV :

```powershell
hdfs dfs -mkdir -p /user/formation/data
hdfs dfs -put "C:\Users\formation\Documents\Big-data-open-data-course-main\donnees\effectifs_vp_bts.csv" /user/formation/data/
hdfs dfs -ls /user/formation/data
```

Astuce : depuis la racine du dépôt, vous pouvez utiliser un chemin relatif converti, par exemple :

```powershell
hdfs dfs -put .\donnees\effectifs_vp_bts.csv /user/formation/data/
```

Lire le fichier depuis HDFS :

```powershell
hdfs dfs -cat /user/formation/data/effectifs_vp_bts.csv | more
```

## 3. Inspecter les blocs et la réplication

```powershell
hdfs fsck /user/formation/data/effectifs_vp_bts.csv -files -blocks
```

Ouvrir **http://localhost:9870** → menu *Utilities* → *Browse the file system* pour parcourir `/user/formation/data` et visualiser blocs et nœuds.

## 4. Tolérance aux pannes (démonstration)

Sous contrôle du formateur : arrêter un DataNode (`stop-dfs.cmd` partiel ou arrêt du service concerné), vérifier que la **lecture** du fichier sur HDFS fonctionne toujours grâce à la **réplication**, puis redémarrer le cluster.

## 5. Synthèse à noter

- Rôle du **NameNode** vs **DataNode**
- Taille des blocs et intérêt pour de gros fichiers
- Lien avec le TP Spark (traitement **au-dessus** de fichiers HDFS ou locaux)

# TP 2.2 — Démonstration Hadoop / HDFSCe notebook **guide** la manipulation HDFS (souvent dans un conteneur Docker). Les commandes ne s’exécutent pas toutes depuis Python : utiliser le terminal du formateur ou les cellules `!` **si** l’environnement du notebook a accès au client `hdfs` et au cluster.**Donnée volumineuse conseillée** : effectifs VP/BTS — voir `Docs/sources_data.md` section TP 2.2.**Références** : `README.md` (section Hadoop).

## 1. Démarrer le cluster (exemple)L’image exacte dépend du formateur. Schéma type :```bash# Exemple indicatif — remplacer par le compose / run fourni en formationdocker compose up -d```Vérifier les conteneurs : `docker ps`.

## 2. Copier un fichier local vers HDFSAdapter le chemin HDFS et le fichier (ex. effectifs déposés dans le conteneur ou monté en volume).```bashhdfs dfs -mkdir -p /user/$USER/datahdfs dfs -put /chemin/vers/effectifs_vp_bts.csv /user/$USER/data/hdfs dfs -ls /user/$USER/data```

## 3. Inspecter les blocs et la réplication```bashhdfs fsck /user/$USER/data/effectifs_vp_bts.csv -files -blocks```Ouvrir l’interface web du NameNode (souvent **http://localhost:9870** selon l’image) pour visualiser blocs et nœuds.

## 4. Tolérance aux pannes (démonstration)Sous contrôle du formateur : arrêter un DataNode, vérifier que la lecture du fichier sur HDFS fonctionne toujours grâce à la réplication.

## 5. Synthèse à noter- Rôle du **NameNode** vs **DataNode**- Taille des blocs et intérêt pour de gros fichiers- Lien avec le TP Spark (traitement **au-dessus** de fichiers HDFS ou locaux)
